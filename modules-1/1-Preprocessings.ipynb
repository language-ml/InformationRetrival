# Load Poems

```python
import sys
import codecs
import tqdm
sys.path.append('../src/util/')
```

```python
all_poet_files = ['../data/farsi/poets/moulavi.txt','../data/farsi/poets/ferdousi.txt','../data/farsi/poets/hafez.txt','../data/farsi/poets/roodaki.txt','../data/farsi/poets/saadi.txt','../data/farsi/poets/shahriar.txt']
```

```python
all_poems_by_poet = dict([(poet_file.split('/')[-1].split('.')[0],[l.strip() for l in codecs.open(poet_file,'r','utf-8').readlines() if len(l.strip())>0][2::]) for poet_file in all_poet_files])
```

# Normalization

```python
from __future__ import unicode_literals
from hazm import *
import random

normalizer = Normalizer()


mesra_collection = all_poems_by_poet['hafez']
mesra_normalized = [[normalizer.normalize(y) for y in x.split()] for x in tqdm.tqdm(mesra_collection)]
```

```python
normalizer.normalize('')
# میوهٔ
```

```python
for x in random.sample(mesra_normalized, 30):
    print(' '.join(x))
```

# Sentence tokenization

```python
mesra_sentences = [sent_tokenize(' '.join(x)) for x in tqdm.tqdm(mesra_normalized)]
```

```python
mesra_sentences
```

# Tokenization

```python
mesra_tokens = [[word_tokenize(sent) for sent in sents] for sents in tqdm.tqdm(mesra_sentences)]
```

# Frequency Analysis

```python
from nltk import FreqDist
import itertools
import pandas as pd

all_tokens = dict()
dataframe = {}

for poet, hemistiches in tqdm.tqdm(all_poems_by_poet.items()):
    all_tokens = list(itertools.chain(*[word_tokenize(hemistich) for hemistich in hemistiches]))
    dataframe[poet] = FreqDist(all_tokens).most_common(50)

freq_analysis = pd.DataFrame(dataframe)    
```

```python
freq_analysis
```

# Stop-words

```python
# Persian Stopwords
# https://github.com/sobhe/hazm/blob/master/hazm/data/stopwords.dat
stopwords = [normalizer.normalize(x.strip()) for x in codecs.open('../data/farsi/stopwords.txt','r','utf-8').readlines()]
```

```python
all_tokens = dict()
dataframe = {}

for poet, hemistiches in tqdm.tqdm(all_poems_by_poet.items()):
    all_tokens = list(itertools.chain(*[[t for t in word_tokenize(hemistich) if t not in stopwords] for hemistich in hemistiches]))
    dataframe[poet] = FreqDist(all_tokens).most_common(50)

freq_analysis = pd.DataFrame(dataframe)    
```

```python
freq_analysis
```
